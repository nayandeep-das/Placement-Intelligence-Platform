# Placement Intelligence Platform

## Project Overview

The Placement Intelligence Platform (PIP) is an ML-powered placement prediction and recommendation system designed to simulate a real-world campus placement process. It combines database management, analytics, machine learning, recommendation systems, and interactive visualization to help students assess placement readiness and identify suitable companies.

## Objectives

- Generate realistic company data
- Populate CompanyRoles table
- Define role-specific technical skills
- Define company eligibility criteria
- Generate realistic student profiles
- Assign technical skills to students
- Populate placement records
- Store all generated data in MySQL

## Why Generate Synthetic Data?

Real placement datasets containing student information, company requirements, salaries, and placement outcomes are generally confidential.

To overcome this limitation, this project generates a synthetic dataset that mimics real campus placement patterns while preserving realistic relationships between students, companies, skills, and eligibility criteria.

The generated dataset is designed to resemble the placement ecosystem at NIT Durgapur and supports the development of machine learning models and recommendation systems.

## Data Generation Strategy

The dataset generation process follows these principles:

- Different branches possess different technical skills.
- Different companies recruit for different job roles.
- Every job role requires a unique set of technical skills.
- Companies define branch eligibility and minimum CGPA requirements.
- Student profiles are generated using realistic academic distributions.
- Placement decisions depend on multiple factors rather than random assignment.

## Database Tables Populated

This notebook generates synthetic data and populates all database tables required for the Placement Intelligence Platform.

### Master (Reference) Tables

These tables store static reference data that is used throughout the platform.

- Branches
- Roles
- Skills

---

### Relationship (Junction) Tables

These tables establish many-to-many relationships between entities and support recommendation, eligibility checking, and skill mapping.

- CompanyRoles
- RoleSkills
- StudentSkills
- CompanyEligibility

---

### Transactional Tables

These tables store dynamic operational data generated during the placement process.

- Students
- Companies
- Placements

## Convert to RoleSkills Dataset

````python
missing = set()

for role, skills in role_skill_map.items():
    for skill in skills:
        if skill not in skill_dict:
            missing.add(skill)

if missing:
    print("❌ Missing Skills:")
    print(sorted(missing))
else:
    print("✅ All skills exist.")
````

````python
missing_roles = set()

for role in role_skill_map.keys():
    if role not in role_dict:
        missing_roles.add(role)

if missing_roles:
    print("❌ Missing Roles:")
    print(sorted(missing_roles))
else:
    print("✅ All roles exist.")
````

````python
role_skills = []

for role, skills_required in role_skill_map.items():

    role_id = role_dict[role]

    for skill in skills_required:

        skill_id = skill_dict[skill]

        role_skills.append(
            (
                role_id,
                skill_id
            )
        )

print(f"Total Role-Skill mappings : {len(role_skills)}")
````

## Reading Required Tables

````python
students_db = pd.read_sql("""
SELECT *
FROM Students
""", conn)

company_roles_db = pd.read_sql("""
SELECT *
FROM CompanyRoles
""", conn)

company_eligibility_db = pd.read_sql("""
SELECT *
FROM CompanyEligibility
""", conn)
````

## Eligibility Lookup

````python
eligibility_dict = {}

for _, row in company_eligibility_db.iterrows():

    company_role = int(row["CompanyRoleID"])
    branch = int(row["BranchID"])
    cgpa = float(row["MinimumCGPA"])

    eligibility_dict.setdefault(company_role, []).append(
        (branch, cgpa)
    )
````

## CompanyRole Lookup by Preferred Role

````python
role_company_dict = (
    company_roles_db
    .groupby("RoleID")["CompanyRoleID"]
    .apply(list)
    .to_dict()
)
````

## Merging Skill Count

````python
skill_count_df = (
    student_skills_df
    .groupby("StudentID")
    .size()
    .reset_index(name="SkillCount")
)

students_db = students_db.merge(
    skill_count_df,
    on="StudentID",
    how="left"
)

students_db["SkillCount"] = (
    students_db["SkillCount"]
    .fillna(0)
    .astype(int)
)

students_db.head()
````

## Placement Score Function

````python
def calculate_score(student):

    score = (
        
    (student["CGPA"] / 10) * 27 +
    (student["CodingScore"] / 100) * 22 +
    (student["AptitudeScore"] / 100) * 14 +
    (student["CommunicationScore"] / 100) * 10 +
    (min(student["DSASolved"], 500) / 500) * 10 +
    (min(student["Projects"], 5) / 5) * 5 +
    (min(student["Internships"], 2) / 2) * 4 +
    (min(student["SkillCount"], 12) / 12) * 3
    
    )

    return round(score, 2)
````

````python
calculate_score(students_db.iloc[0])
````

## Placement Probability Function

````python
def placement_probability(score):

    if score >= 90:
        return 0.98
    elif score >= 80:
        return 0.90
    elif score >= 70:
        return 0.75
    elif score >= 60:
        return 0.55
    elif score >= 50:
        return 0.35
    else:
        return 0.15
````

````python
placement_probability(82)
````

## Placement Filling Loop

````python
placements = []
placed_student_ids = []

for _, student in students_db.iterrows():

    student_id = student["StudentID"]
    role_id = student["PreferredRoleID"]
    branch_id = student["BranchID"]
    cgpa = student["CGPA"]

    
    company_roles = role_company_dict.get(role_id, [])

    eligible_company_roles = []

    
    for company_role in company_roles:

        rules = eligibility_dict.get(company_role, [])

        for eligible_branch, cutoff in rules:

            if branch_id == eligible_branch and cgpa >= cutoff:
                eligible_company_roles.append(company_role)
                break

    
    if not eligible_company_roles:
        continue

    
    score = calculate_score(student)

    
    probability = placement_probability(score)

    
    if random.random() < probability:

        selected_company_role = random.choice(eligible_company_roles)

        placements.append(
            (
                student_id,
                selected_company_role,
                "Placed",
                date.today()
            )
        )

        placed_student_ids.append(student_id)
````
