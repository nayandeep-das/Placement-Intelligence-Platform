## Transaction Tables

Unlike the master tables, transaction tables store dynamic information such as students, companies, placements, and skill mappings. These tables reference the master tables using foreign keys to maintain data integrity.

## Students Table

The **Students** table stores the academic, technical, and personal profile of every student in the Placement Intelligence Platform. It acts as the central entity of the database and is linked to branches, skills, and placement records through foreign key relationships.

**Purpose**
- Store student profile information.
- Maintain academic and technical performance data.
- Serve as the primary source for placement prediction and recommendation.

## Companies Table

The **Companies** table stores information about organizations participating in the placement process. Each company can offer multiple job roles with different eligibility criteria.

**Purpose**
- Store company information.
- Maintain records of recruiting organizations.
- Link companies with offered job roles and placement records.

## CompanyRoles Table

The **CompanyRoles** table represents the relationship between companies and the job roles they offer. Since a company can offer multiple roles and similar roles can be offered by multiple companies, this table resolves the many-to-many relationship.

**Purpose**
- Map companies to their offered job roles.
- Maintain role-specific information for each company.
- Serve as a reference for eligibility criteria and recommendations.

## StudentSkills Table

The **StudentSkills** table stores the technical and soft skills possessed by each student. Since a student can possess multiple skills and a skill can belong to multiple students, this table represents a many-to-many relationship.

**Purpose**
- Map students to their acquired skills.
- Support recommendation and skill-gap analysis.
- Calculate the student's skill count and technical profile.

## RoleSkills Table

The **RoleSkills** table stores the skills required for each job role. It represents a many-to-many relationship between job roles and skills, enabling role-based recommendation and skill matching.

**Purpose**
- Define required skills for each job role.
- Compare student skills with industry requirements.
- Support career recommendation and missing skill identification.

## CompanyEligibility Table

The **CompanyEligibility** table stores the minimum eligibility criteria required by companies for specific job roles. These criteria include academic performance and other recruitment requirements.

**Purpose**
- Store company-specific eligibility criteria.
- Determine whether a student satisfies placement requirements.
- Support company recommendation and eligibility analysis.

## Placements Table

The **Placements** table stores the final placement outcome of each student. It records whether a student was placed along with the associated company and offered role.

**Purpose**
- Store placement results.
- Track successful placements.
- Provide target labels for machine learning model training and evaluation.
